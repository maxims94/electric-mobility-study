# Dataset 2: Exploration 

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('../data/dataset2.csv',index_col='country')
```

## Plot EV share against various characteristics

```python
g = sns.PairGrid(df,x_vars=df.drop('max_ev_p',axis=1).columns, y_vars=['max_ev_p'])
g.map(sns.regplot)
```

## Correlations between EV market share and various characteristics 

```python
print(df.corr().loc['max_ev_p'])
```

## Observations 

You mostly see the trends that are to be expected:
* EV market is positively correlated with same-sex marriage
* The higher the gini index, the greater income inequality, the lower the market share of EVs
* A greater GDP (PPP) per capita is related to a higher share of EVs
* A higher social progress index is also positively correlated with EVs
* It seems that EV share is independent of election results of nationalist parties
    * This is a surprise, as one might expect them to be slow down policies on climate change, which would result in less subsidies for EV and hence, a lower market share
